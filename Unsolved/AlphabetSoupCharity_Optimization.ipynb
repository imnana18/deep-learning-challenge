```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd 
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head(50)
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df = application_df.drop(columns = ['EIN', 'NAME'])
```

```python
# Determine the number of unique values in each column.
print(application_df.nunique())
```

```python
# Look at APPLICATION_TYPE value counts for binning
app_type_value_counts = application_df['APPLICATION_TYPE'].value_counts()
app_type_value_counts
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
cutoff_val = 1000
application_types_to_replace = [app_type for app_type, count in app_type_value_counts.items() if count <= cutoff_val]

# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
classification_value_counts = application_df['CLASSIFICATION'].value_counts()
classification_value_counts[:50]
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
counts_greater_than_1 = classification_value_counts[classification_value_counts > 1]
print(counts_greater_than_1)
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
cutoff_val2 = 100
classifications_to_replace = [cls_type for cls_type, count in classification_value_counts.items() if count <= cutoff_val2]

# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")
    
# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
# Bin 'ASK_AMT' column

# Check the number range of the column
print(application_df['ASK_AMT'].min())
print(application_df['ASK_AMT'].max())

# Define the bin edges
bins =  [0,  1000,   10000,   100000,   1000000,   10000000,    100000000, 1000000000, 10000000000]
labels = ['0-1000', '1000-10000', '10000-100000', '100000-1000000', '1000000-10000000', '10000000-100000000', '100000000-1000000000', '1000000000-10000000000']

application_df['ASK_AMT'] = pd.cut(application_df['ASK_AMT'], bins, labels = labels)

application_df.head()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
columns_to_encode = ['AFFILIATION', 'USE_CASE', 'ORGANIZATION', 'APPLICATION_TYPE', 'CLASSIFICATION', 'INCOME_AMT','ASK_AMT']
enc_app_df = pd.get_dummies(application_df, columns = columns_to_encode, dtype = int)
enc_app_df.head()
```

```python
# Split our preprocessed data into our features and target arrays
y = enc_app_df['IS_SUCCESSFUL'].values
X = enc_app_df.drop(columns = ['IS_SUCCESSFUL', 'STATUS', 'SPECIAL_CONSIDERATIONS']).values

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 71, stratify = y)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
number_input_features = len(X_train[0])
print(number_input_features)
hidden_nodes_layer1 = 32
hidden_nodes_layer2 = 16
# hidden_nodes_layer3 = 8

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense
    (
    units = hidden_nodes_layer1, 
    activation = 'relu', 
    input_dim = number_input_features
    )
)

#nn.add(tf.keras.layers.Dropout(.2))

# Second hidden layer
nn.add(tf.keras.layers.Dense
    (
    units = hidden_nodes_layer2, 
    activation = 'sigmoid'
    )
)
#nn.add(tf.keras.layers.BatchNormalization(synchronized=True))

# Third hidden layer
# nn.add(tf.keras.layers.Dense
#     (
#     units = hidden_nodes_layer3, 
#     activation = 'relu'
#     )
# )

# Output layer
nn.add(tf.keras.layers.Dense
    (
    units = 1, 
    activation = 'sigmoid'
    )
)

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
```

```python
# Train the model
model_fit = nn.fit(X_train_scaled, y_train, epochs = 50, batch_size=8)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled, y_test, verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
